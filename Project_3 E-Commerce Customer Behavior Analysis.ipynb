#### Analyze Sales Patterns

Look for trends such as:

Which country buys the most?

Which product is most sold?

Which customer spends the most?

```python
most_country_buys = data.groupby('Country')['TotalPrice'].count().sort_values(ascending=False).reset_index()
most_country_buys.head(20)
```

```python
most_sold_product = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).reset_index()
most_sold_product.head(20)
```

```python
customer_spend_most = data.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).reset_index()
customer_spend_most.head(20)
```

## Step 4: Exploratory Data Analysis (Visual Insights)

```python
# Top Selling Products
plt.figure(figsize=(10,5))
sns.barplot(data=most_sold_product.head(10), x='Quantity', y='Description', palette='viridis')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Description")
plt.show()

```

```python
# Sales by Country

plt.figure(figsize=(10,5))
sns.barplot(data = most_country_buys.head(10), x='TotalPrice', y='Country', palette='viridis')
plt.title("Top 10 Countries by Total Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Country")
plt.show()
```

#### 3. Monthly Revenue Trend
Convert your invoice date column to a datetime type.

Extract the month-year and group by it.

Plot a line chart to show how sales changed over time.

```python
Month_revenue_trend = data.groupby(['Year', 'Month'])['TotalPrice'].sum().reset_index()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=Month_revenue_trend, x='Month', y='TotalPrice', hue='Year', marker='o')
plt.title("Monthly Revenue Trend by Year")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

### 4. Hourly Sales Pattern (Peak hours)

Extract the hour from your invoice time column.

Group by hour, sum the total sales, and make a line or bar chart showing which hours have the most sales.

```python
data['Hours'] = pd.to_datetime(data['InvoiceDate']).dt.hour
data
```

```python
Hourly_Sales_Pattern = data.groupby('Hours')['TotalPrice'].sum().reset_index().sort_values('Hours')
Hourly_Sales_Pattern
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=Hourly_Sales_Pattern, x='Hours', y='TotalPrice', marker='o', color='purple')
plt.title("Hourly Revenue Trend by Year")
plt.xlabel("Hour of the Day (0–23)")
plt.ylabel("Total Revenue")
plt.grid(True, alpha=0.3)
plt.show()
```

### Project Summary & Key Insights — E-Commerce Customer Behavior Analysis

Top revenue and products

The dataset shows that the largest revenue comes from the United Kingdom (by far). This is common for the UK Online Retail dataset — UK contributes the majority of transactions and revenue.

The best-selling items include popular gift items such as WHITE HANGING HEART T-LIGHT HOLDER and several home-decor items (these appear repeatedly in the top-sold product list). These SKUs drive a large share of quantity sold.

Customer purchasing behavior

A small set of loyal customers contribute disproportionately to total revenue — the top customers by TotalPrice account for a notable share of sales.

Most transactions involve small quantities of single items, but some invoices include many line items (useful for bundling or cross-sell strategies).

Time patterns

Monthly revenue shows strong seasonality with a clear peak toward November / December (holiday season), and a big increase in late 2011 in your plotted data. This suggests promotional and seasonal effects.

Hourly sales pattern indicates higher sales during daytime hours (typical peaks occur late morning to afternoon), which helps schedule marketing pushes and staffing for order fulfilment.

Business recommendations

Focus marketing and promotions on the UK and on the top-selling product categories during October–December to maximize holiday revenue.

Identify top customers for loyalty programs — consider targeted offers to increase their lifetime value.

Consider product bundles combining high-frequency small items with top revenue-driving items to increase average order value (AOV).

Limitations

The dataset is heavily skewed to the UK. If expanding analysis to global markets, a more geographically balanced dataset is needed.

The invoice-level granularity requires careful aggregation (counting InvoiceNo vs summing Quantity / TotalPrice) depending on the business question.
